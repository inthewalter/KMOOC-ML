자, 그러면, 이러한 자료에 대한 기초지식과 강의를 통해 배운 것으로 다음과 같은 붓꽃 자료에 대한 이진 분류 코드를 실험해보길 바랍니다. 

[참고사항]  joy 모듈은 5-1 혹은 5-2 강의 영상의 하단에 있는 공지사항을 숙지하시고, [학습자료 다운로드]에서 joyML.zip를 다운받아 설치하시면 사용할 수 있습니다. 


다음 코드는 붓꽃 데이터셋을 퍼셉트론으로 이진 분류하는 것입니다.   
다음과 같은 결과를 얻기 위하여 코드를 완성하기 위한 적절한 하이퍼파라미터를 선택하십시오.



```python
import joy
import matplotlib.pyplot as plt
%matplotlib inline

X, y = joy.iris_data()

# ppn = joy.Perceptron(eta = ?, epochs = ?)
ppn = joy.Perceptron(eta = 0.01, epochs = 10)
ppn.fit(X, y)
plt.plot(range(1, len(ppn.cost_) + 1), ppn.cost_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of updates')
plt.show()

# joy.plot_xyw(X, y, ?)
joy.plot_xyw(X, y, ppn.w)
```
