# 6-3 다층 퍼셉트론






퍼셉트론이란 인공 신경망에 연결된 하나의 인공 뉴런을 말한다.  
이런 뉴런이 학습할 수 있도록 만든 퍼셉트론 알고리즘을 로젠 블라트가 처음으로 제안했다.  
그런데 하나의 퍼셉트론으로는 풀 수 없는 문제가 있었다.



바로 배타적 논리합(XOR) 문제인데  
선형 분류만을 할 수 있는 퍼셉트론은 이 간단한 XOR 문제를 풀 수 없었다.

나중에 펄 워보스가 발표한 역전파를 이용한 다층 퍼셉트론으로 XOR문제를 풀 수 있었는데,   
다층 퍼셉트론은 영어로 MLP(Multi-Layer Perceptron)이라고 한다.

퍼셉트론을 여러 층으로 연결한 구조이다.

간단한 다층 구조이지만, 각 층마다 우리가 퍼셉트론의 수를 얼마든지 증가시킬 수 있다.  

입력으로 구성된 층을 입력층이라 한다. (input layer)  
출력으로 구성된 층을 출력층, (output layer)  

입력층과 출력층 사이에 있는 층을 은닉층 (hidden layer)라고 한다.  

입력과 출력 사이에 감추어져 있어서 그렇게 이름이 지어졌고,  
은닉층의 수가 증가하게 되면, 그게 바로 Deep Neural Network가 생겨나는 것이다.

이와 같은 구조를 2층 혹은 3층 인공 신경망이라고 하는데,  
입력층은 입력을 받아 단순히 전달하는 일만 하기 때문에 층으로 간주하지 않는 사람들도 있다.
