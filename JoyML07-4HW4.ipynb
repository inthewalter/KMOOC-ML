## joyHW 4-1

  3   1 2   ,      ?  
  (ReLU) 




```python
import numpy as np

#     .
def relu(x):
    return np.maximum(0, x)

x = np.array([[1], [0]])
W1 = np.array([[0.1, 0.9], [0.5, 0.7], [0.3, 0.7]])
Z1 = np.dot(W1, x)
A1 = relu(Z1)

W2 = np.array([[0.1, 0.9, 0.3], [0.5, 0.7, 0.2]])
Z2 = np.dot(W2, A1)
yhat = relu(Z2)

print(yhat)
```

```python
import numpy as np

#     .
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.array([[1], [0]])

W1 = np.array([[0.1, 0.0], [-0.5, 0.7], [-0.3, 0.7]])
Z1 = np.dot(W1, x)
A1 = sigmoid(Z1)

W2 = np.array([[0.4, 0.5, -0.2], [-0.1, 0.0, -0.3], [-0.5, 0.7, -0.2]])
Z2 = np.dot(W2, A1)
A2 = sigmoid(Z2)

W3 = np.array([[-0.1, 0.0, -0.3], [-0.5, 0.7, 0.2]])
Z3 = np.dot(W3, A2)
A3 = sigmoid(Z3)

yhat = A3

print(yhat)
print('\n\n\n\n')

g = lambda x : 1 / (1 + np.exp(-x)) # sigmoid activation functin

x = np.array([1, 0])

W1 = np.array([[0.1, 0.0], [-0.5, 0.7], [-0.3, 0.7]])
z1 = np.dot(W1, x)
a1 = g(z1)

W2 = np.array([[0.4, 0.5, -0.2], [-0.1, 0.0, -0.3], [-0.5, 0.7, -0.2]])
z2 = np.dot(W2, a1)
a2 = g(z2)

W3 = np.array([[-0.1, 0.0, -0.3], [-0.5, 0.7, -0.2]])
z3 = np.dot(W3, a2)
yhat = g(z3)

print("z1:", z1)
print("a1:", a1)
print("z2:", z2)
print("a2:", a2)
print("z3:", z3)
print("yhat:", yhat)
```
