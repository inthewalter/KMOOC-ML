지금까지 우리는 세 층으로 이루어진 신경망을 다루었습니다. 세 층으로 이루어진 뉴론들이 다음 층이나 이전 층의 다른 모든 뉴론들과 연결되어 있는 것이 복잡해 보입니다. 

신호가 입력층으로 들어오고, 여러 뉴론을 지나서 출력이 되는 것이 복잡해 보이지만, 사실상 따지고 보면 따라가기가 복잡한 곱셈과 덧셈에 불과합니다.   
물론, 우리가  이런 계산을 모두 코딩을 처리하겠지만, 그 전에 우리는 이런 계산이 어떻게 일어나는 확실히 알아야 합니다.

우선, 신경망 내부에서 실제로 어떤 일이 일어나는지에 대해 이해해야 합니다.   
순방향 신경망의 신호가 어떻게 처리되는지 배웠는데, 이제 신경망에서 일어나는 일들을 추적하며 신경망의 신호처리를 계산해보는 것이 이번 과제입니다. 

강의에서 언급했듯이, 다음과 같은 신경망의 표기법에 익숙해지는 것이 필수적입니다. 



이렇게 입력층에서 출발하여 은닉층들을 거쳐서 출력층에서 결과를 얻는 것을 순방향 신경망(feed-forward  neural network)이라고 부릅니다. 

여기서 사용하는 두 활성화 함수는 렐루(ReLU, Rectified Linear Unit)함수와 시그모이드(Sigmoid)함수입니다.



결론적으로, 순방향 신경망이 어떻게 신호를 처리하는지 이번 과제에서 계산해 보자는 것입니다.   
이 과제는 여러분이 앞으로 기계학습을 공부하는데 꼭 통과해야 할 관문과 같은 필수적 과제이니, 직접 계산해보고 지나가는 것이 좋겠습니다. 

## joyHW 4-1

다음과 같이 3층 신경망의 은닉층 1과 2의 가중치와 입력이 주어질때, 출력으로 나오게 되는 의 결과는 무엇인가요?  
활성화 함수는 렐루(ReLU)함수를 사용합니다




```python
import numpy as np

# 활성화 함수로 렐루 함수를 사용한다.
def relu(x):
    return np.maximum(0, x)

x = np.array([[1], [0]])
W1 = np.array([[0.1, 0.9], [0.5, 0.7], [0.3, 0.7]])
Z1 = np.dot(W1, x)
A1 = relu(Z1)

W2 = np.array([[0.1, 0.9, 0.3], [0.5, 0.7, 0.2]])
Z2 = np.dot(W2, A1)
yhat = relu(Z2)

print(yhat)
```

## joyHW 4-2

조금 더 복잡한 신경망을 구현해봅니다. 은닉층이 2개 있으며, 각각의 은닉층이 3개의 뉴런을 사용하며, 시그모이드 활성화 함수를 사용하는 신경망이 있다고 합시다. 아래와 같이 가중치를 주고, 입력값이 들어왔을 때, 신경망이 예측한 값 을 계산하세요. 결과값으로 다음 보기중에 가장 가까운 값을 고르세요.



```python
import numpy as np

# 활성화 함수로 시그모이드 함수를 사용한다.
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.array([[1], [0]])

W1 = np.array([[0.1, 0.0], [-0.5, 0.7], [-0.3, 0.7]])
Z1 = np.dot(W1, x)
A1 = sigmoid(Z1)

W2 = np.array([[0.4, 0.5, -0.2], [-0.1, 0.0, -0.3], [-0.5, 0.7, -0.2]])
Z2 = np.dot(W2, A1)
A2 = sigmoid(Z2)

W3 = np.array([[-0.1, 0.0, -0.3], [-0.5, 0.7, 0.2]])
Z3 = np.dot(W3, A2)
A3 = sigmoid(Z3)

yhat = A3

print(yhat)
print('\n\n\n\n')

g = lambda x : 1 / (1 + np.exp(-x)) # sigmoid activation functin

x = np.array([1, 0])

W1 = np.array([[0.1, 0.0], [-0.5, 0.7], [-0.3, 0.7]])
z1 = np.dot(W1, x)
a1 = g(z1)

W2 = np.array([[0.4, 0.5, -0.2], [-0.1, 0.0, -0.3], [-0.5, 0.7, -0.2]])
z2 = np.dot(W2, a1)
a2 = g(z2)

W3 = np.array([[-0.1, 0.0, -0.3], [-0.5, 0.7, -0.2]])
z3 = np.dot(W3, a2)
yhat = g(z3)

print("z1:", z1)
print("a1:", a1)
print("z2:", z2)
print("a2:", a2)
print("z3:", z3)
print("yhat:", yhat)
```
